```python
#Renames columns 
d = {"BUNDESLAND_NAME":"Bundesland","STATIONSNAME":"Station","STATIONS_ID":"ID",
     "STATIONSHÖHE IN METER ÜBER NN":"StationsHöhe"
    ,"NÖRDLICHE GEOGRAFISCHE BREITE IN GRAD":"GeoLänge","ÖSTLICHE GEOGRAFISCHE LÄNGE IN GRAD":"GeoBreite",
     "JAHR":"Jahr","MONAT":"Monat","TAG":"Tag","TAGESSUMME DER NIEDERSCHLAGSHÖHE IN MM":"NN_mm",
     "Anzahl der Tage > 1,0 mm":"NN_1mm","TAGESMITTEL DER WINDGESCHWINDIGKEIT IN m/s":"W_ms",
     "Anzahl der Tage > 5,3 m/s":"W_5.3ms"}

Niederschlag.rename(columns= d,inplace=True)

Niederschlag
```

```python
#Convert the columns Jahr, Monat and Tag to single distinguishable timestamps

Niederschlag= Niederschlag.astype({"Jahr":str,"Monat":str,"Tag":str})

Niederschlag["Datum"]=Niederschlag["Jahr"]+"-"+Niederschlag["Monat"]+"-"+ Niederschlag["Tag"]

Niederschlag["Datum"]= Niederschlag["Datum"].apply(pd.Timestamp)

#Remove no longer needed columns
Niederschlag=Niederschlag.drop(labels=["Jahr","Tag","Monat"], axis=1)
Niederschlag
```

```python
Alle_BL =Niederschlag.groupby(by=["Bundesland", "Station"])
Alle_BL
```

```python
 Alle_BL.sum()["NN_mm"].hist(bins=20)
plt.title(label="Verteiliung der Summe Jährliche Niederschlag pro Station")
```

```python
Station =Alle_BL.get_group(("Baden-Württemberg",'Freiburg')).copy()
```

```python
Station["NN_cumu"]= Station["NN_mm"].cumsum()
```

```python
x=Station["Datum"]
y=Station["NN_cumu"]
plt.plot(x,y)
plt.title(label="Cumulative Niederschlag in Wetterstation Freiburg")
```

```python

x= Niederschlag["Datum"]
y = Niederschlag["NN_mm"]

s= 2*y
plt.scatter(x, y, s=s,  alpha=0.5)
plt.title(label="Unübersichtlich")
```

```python
x= Niederschlag["Datum"]
y = Niederschlag["NN_mm"]


#Große der Grafik
plt.figure(figsize=(12,8 ), dpi=80)
plt.scatter(x, y, s=y,  alpha=0.3)
plt.title("Besser dargestellt", fontsize=14)
```

```python
x= Niederschlag["Datum"]
y = Niederschlag["NN_mm"]

plt.figure(figsize=(12,8 ), dpi=80)
plt.ylim(top=110)
sns.kdeplot( x=x,y=y, cmap="Reds", shade=True)
plt.title("Density Map of Rain Events in 2020", loc="center",fontsize=14)
plt.show()
print("Und noch viel besser")
```
